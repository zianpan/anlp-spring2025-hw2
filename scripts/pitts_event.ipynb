```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from tqdm import tqdm

# Set up the WebDriver
# driver_path = "/path/to/chromedriver"  # Replace with the correct path
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
# Open the website
url = "https://pittsburgh.events/"
driver.get(url)
time.sleep(3)  # Allow page to load

# Keep clicking "Show More" until it's no longer available
for i in range(30):  # Adjust the range if necessary
    try:
        show_more_button = driver.find_element(By.CLASS_NAME, "ldm")  # Target the class name of the button
        driver.execute_script("arguments[0].click();", show_more_button)  # Click using JS
        time.sleep(2)  # Wait for new content to load
    except:
        print("No more 'Show More' button found.")
        break

# Extract all event data
events = driver.find_elements(By.CLASS_NAME, "date-row")  # Adjust class name if necessary

# event_list = []

# Close the WebDriver
# driver.quit()

# # Print or save results
# import pandas as pd
# df = pd.DataFrame(event_list)
# df.to_csv("pittsburgh_events.csv", index=False)
# print(df)
```

```python
event_data = []
for event in tqdm(events):
    elements = event.find_elements(By.XPATH, ".//*")  # Get all child elements
    event_info = {}
    
    for elem in elements:
        class_name = elem.get_attribute("class")
        text = elem.text.strip()
        
        if class_name and text:  # Store only non-empty text
            event_info[class_name] = text

    event_data.append(event_info)
```

```python
driver.quit()
```

```python
s =  ''
```

```python
for e in event_data:
    for k, v in e.items():
        # print(v)?
        v = v.replace('\n', ' ')
        s += f"{k}:{v}  "
        
    s += "\n"
```

```python
with open('../data/pitts_event/pittsburgh_events.txt', 'w') as f:
    f.write(s)
```
